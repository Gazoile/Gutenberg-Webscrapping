```python
import requests
livre = requests.get('https://www.gutenberg.org/cache/epub/17989/pg17989-images.html')
#récuperer le contenu html d'une page web
livre.text[:100]
#import de la librarie bs4 et du module beautifulsoup pour mettre en forme le contenu qu'on a récup
from bs4 import BeautifulSoup
#je crée un objet qui contient la page sous forme d'arbre ("html tree")  avec toutes les balises html de la page
soup = BeautifulSoup(livre.content,"html.parser")
#affichage d'une balise nommé <h1> test
soup.h1
#je récupère tout le contenu de ma page web html sous forme de texte (sans balises)
text = soup.get_text()
#j'affiche mon texte et sa longueur
text
len(text)


# essayer d'épurer l'affichage de notre texte à l'aide de ... ?
```

```python
#récupérer le premier chapitre

soup.find(class_='chapter')
chapitre1 = soup.find(class_='chapter')
type(chapitre1)
```

```python
#récupérer tous les chapitres
soup.find_all(class_='chapter')
all_chapter = soup.find_all(class_='chapter')
#retourner le chapitre 1 de all texte avec [0] afin de l'épurer avec la fonction text
type(all_chapter)

print(type(all_chapter[0]))
## len(all_chapter[0]).text. ne fonctionne pas car .text s'applique à len all chapter [0]
## qui est un entier
# ici ça fonctionne on a notre nombre de caractère
len(all_chapter[0].text)

#Récupperer tout les chapitres de notre texte
Le_texte = []
for i in all_chapter:
  Le_texte.append(i.text)

compt = 0
#1)définition de char à remplacé
delete0 = "«,:;?!-()1234567890—».\n-\r"
#création d'une boucle qui parcours les chapitres du texte
for x in range(len(Le_texte)):
  #création d'une boucle qui parcours les chars à remplacé
  for i in delete0:
    #utilisation de la method replace sur chacun des chapitres "[x]"
    #Le_texte[x] est à chaque fois remplacé par ce que renvoie replace
    #sinon le résultat n'est pas stocké et tout les char ne sont pas supprimé
    Le_texte[x]  = Le_texte[x].replace(i," ")
  #Le_texte[x] est mis à jour, une fois sorti de la seconde du boucle car autrement
  #dès la premiere itération le texte[x] va devenir une liste et non plus une string
  #la replace() ne fonctionnera plus
  # je met en miniscule mon texte
  Le_texte[x] = Le_texte[x].lower()
  Le_texte[x] = Le_texte[x].split()[1:]



  compt = compt + len(Le_texte[x])
print(compt)
print(Le_texte)







```

```python

compteur = {}
for chap in Le_texte:
  for mot in chap:
      if mot not in compteur:
        compteur[mot] = 1
      else:
        compteur[mot] += 1
#print(compteur)

L = {}
for v in compteur:
  if len(v) >= 5 :
    L[v] = compteur[v]

```
